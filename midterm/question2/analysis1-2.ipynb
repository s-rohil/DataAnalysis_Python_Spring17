```python
import os
import csv
import json
import datetime
import calendar
import collections
import matplotlib.pyplot as plt
```

```python
directory='../Data'
print(os.path.exists(os.path.join(directory,'Sports')))
```

```python
nyt_dict={}
ap_dict={}
reuters_dict={}
unknown_dict={}
frequency={n: {} for n in range(1, 13)}
popular_sport= {n: {} for n in range(1, 13)}
directory='analysis1-2'
file_name='outputs1'+'.csv'
#print(type(file_name))
if not os.path.exists(directory):
    os.makedirs(directory)
else:
    with open(os.path.join(directory,file_name),'w') as csvfile:
        output = csv.writer(csvfile)
        output.writerow(['Month','Associated Press','Reuters','The New York Times','Unknown source','Popular Sport'])
        for di,subdir,files in os.walk(os.path.join(directory,'../../Data/Sports')):
            #print(os.path.exists(os.path.join(directory,'../../Data/Sports')))
            for file in files:
                if file.endswith(".json"):
                    with open(os.path.join(di,file),'r',encoding='utf-8', errors='ignore') as temp:
                        #print(file)
                        data = json.load(temp)
                        #print(data)
                        #print(file)
                        source= data['source']
                        #print(source)
                        temp_date = data['pub_date']
                        date= datetime.datetime.strptime(temp_date[:10],'%Y-%m-%d').date()
                        sport_name = data['subsection_name']
                        frequency[date.month][sport_name] = frequency[date.month].get(sport_name, 0 ) + 1
                        #print(source)
                        original_source = data['byline']
                        #print(original_source)
                        if source=='AP':
                            #ap_count+=1
                            #temp_date = data['response']['docs'][i]['pub_date']
                            #ap_date = datetime.datetime.strptime(temp_date[:10],'%Y-%m-%d').date()
                            #print(ap_date.month)
                            ap_dict[date.month] = ap_dict.get(date.month,0)+1
                            #print(date.month)
                        elif source=='Reuters':
                            #reuters_count+=1
                            #r_date = datetime.datetime.strptime(temp_date[:10],'%Y-%m-%d').date()
                            #print(r_date.month)
                            reuters_dict[date.month] = reuters_dict.get(date.month,0)+1
                        elif source=='The New York Times':
                            #reuters_count+=1
                            #ny_date = datetime.datetime.strptime(temp_date[:10],'%Y-%m-%d').date()
                            #print(ny_date.month)
                            nyt_dict[date.month] = nyt_dict.get(date.month,0)+1
                            #frequency[date.month]
                        elif source=='':
                            #print(file)
                            if 'ASSOCIATED PRESS' in data['byline']['original']:
                                #print('ap')
                                ap_dict[date.month] = ap_dict.get(date.month,0)+1
                            elif 'REUTERS' in data['byline']['original']:
                                #print('re')
                                reuters_dict[date.month] = reuters_dict.get(date.month,0)+1
                            elif 'NEW YORK TIMES' in data['byline']['original']:
                                #print('nyt')
                                nyt_dict[date.month] = nyt_dict.get(date.month,0)+1
                            else:
                                #print('un')
                                unknown_dict[date.month] = unknown_dict.get(date.month,0)+1
                        
                        
                                
        for i in range(1,13):
            popular_sport[i]= sorted(frequency[i].items(),key = lambda x: x[1] ,reverse=True)                    
                
        for i in range(1,13): #change it to 13 after downloading more data
            #getting the total number of articles printed by various sources sorted by month
            output.writerow([calendar.month_name[i],ap_dict.get(i),reuters_dict.get(i),nyt_dict.get(i),unknown_dict.get(i),popular_sport[i][0][0]])
        
        
                        
                    
                    
                    
                
        
```

```python
ap_count=0
reuters_count=0
nyt_count=0
for key in contributor:
    if 'AP' in contributor[key].keys():
        ap_count+=contributor[key]['AP']
    if 'Reuters' in contributor[key].keys():
        reuters_count+=contributor[key]['Reuters']
    if 'The New York Times' in contributor[key].keys():
        nyt_count+=contributor[key]['The New York Times']
labels=['AP','Reuters','The New York Times']
sizes=[ap_count,reuters_count,nyt_count]
colors=['limegreen','lightskyblue','yellow']
plt.title('Most Important Source',y=1.18)
plt.pie(sizes, colors=colors,labels=labels,autopct="%1.1f%%", shadow=True, startangle=135)
plt.axis('equal')
plt.show()
```
